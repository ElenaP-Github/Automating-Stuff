```python
## import packages

import requests
from bs4 import BeautifulSoup
```

```python
## retrieve the Website URL

URL= 'https://www.monster.com/jobs/search/?q=Data-Analyst&where=New-York__2C-NY'
page= requests.get(URL)#This code performs an HTTP request to the given URL. It retrieves the HTML data that the server sends back and stores that data in a Python object.
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='ResultsContainer')#Find Elements by ID
# print(results.prettify())
```

```python
# Find Elements by HTML Class Name
job_elems = results.find_all('section', class_='card-content')
for job_elem in job_elems:
    print(job_elem, end='\n'*2)
```

```python
# That’s already pretty neat, but there’s still a lot of HTML!

#You’ve seen earlier that your page has descriptive 
#class names on some elements

for job_elem in job_elems:
    # Each job_elem is a new BeautifulSoup object.
    # You can use the same methods on it as you did before.
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')
    if None in (title_elem, company_elem, location_elem):
        continue
    print(title_elem.text)#Extract Text From HTML Elements
    print(company_elem.text)#Extract Text From HTML Elements
    print(location_elem.text)#Extract Text From HTML Elements
    print()
```
